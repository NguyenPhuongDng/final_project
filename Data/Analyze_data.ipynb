Bi·ªÉu ƒë·ªì t·ªïng s·ªë l∆∞·ª£t like, haha, y√™u th√≠ch, ...c·ªßa t·∫•t c·∫£ c√°c b√†i

```python
import ast 
all_reactions = df_full_post['reactions'].apply(ast.literal_eval)
total_reactions = {'like': 0, 'love': 0, 'haha': 0, 'wow': 0, 'care': 0, 'sad': 0, 'angry' :0}

for reactions in all_reactions:
    for reaction, count in reactions.items():
        total_reactions[reaction] += count
```

```python
total_reactions
```

```python
sns.set_theme(style="whitegrid")
sns.set(rc={'figure.figsize':(15,7)})
sns.set(font_scale=1.5)

# plot the reactions
fig, ax = plt.subplots()
ax = sns.barplot(x=list(total_reactions.keys()), y=list(total_reactions.values()))
ax.set_title("Bi·ªÉu ƒë·ªì t·ªïng s·ªë l∆∞·ª£t like, haha, love, ...c·ªßa t·∫•t c·∫£ c√°c b√†i")
ax.set_ylabel("S·ªë l∆∞·ª£ng t∆∞∆°ng t√°c")
ax.set_xlabel("Ki·ªÉu t∆∞∆°ng t√°c")
plt.xticks(rotation=-45)
plt.show()
```

Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán s·ª± t∆∞∆°ng quan gi·ªØa s·ªë l∆∞·ª£t t∆∞∆°ng t√°c v√† th·ªùi gian ƒëƒÉng b√†i trong ng√†y

```python
df_full_post_sorted = df_full_post.sort_values(by='hour_only')

fig, ax1 = plt.subplots(figsize=(30, 20))
values = [x for x in range(df_full_post_sorted.shape[0])]

sns.barplot(x=values, y='reaction_count', data=df_full_post_sorted, ax=ax1, color='blue', label='reaction_count')

ax1.set_xlabel('X')
ax1.set_ylabel('reaction_count', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()

sns.lineplot(x=values, y='hour_only', data=df_full_post_sorted, ax=ax2, color='red', label='hour_only')

ax2.set_ylabel('hour_only', color='red')
ax2.tick_params('y', colors='red')
plt.xticks(rotation=90)
plt.title('Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán s·ª± t∆∞∆°ng quan gi·ªØa s·ªë l∆∞·ª£t t∆∞∆°ng t√°c v√† th·ªùi gian ƒëƒÉng b√†i trong ng√†y')
fig.tight_layout()

plt.show()
```

Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán ·∫£nh h∆∞·ªüng c·ªßa lo·∫°i b√†i ƒëƒÉng t·ªõi l∆∞·ª£t t∆∞∆°ng t√°c

```python
df_vid = df_full_post[df_full_post["video_id"].isna() == False]
df_image = df_full_post[df_full_post["image_id"].isna() == False]
df_text = df_full_post[(df_full_post["image_id"].isna() == True) & (df_full_post["video_id"].isna() == True)]
values = [df_vid["reaction_count"].mean(), df_image["reaction_count"].mean(), df_text["reaction_count"].mean()]
values_count = [df_vid.shape[0], df_image.shape[0], df_text.shape[0]]
categories = ['video', 'image', 'text']

plt.figure(figsize=(12, 6))
#bi·ªÉu ƒë·ªì 1
plt.subplot(1, 2, 1)
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green'])
plt.title('Ph·∫ßn trƒÉm t∆∞∆°ng t√°c trung b√¨nh theo lo·∫°i b√†i ƒëƒÉng', pad=20) 
plt.gca().set_aspect('equal')

#bi·ªÉu ƒë·ªì 2
plt.subplot(1, 2, 2)
plt.bar(categories, values_count)
plt.title('S·ªë b√†i post theo t·ª´ng lo·∫°i', pad=20) 

plt.show()
```

```python
Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán ·∫£nh h∆∞·ªüng c·ªßa th·ªùi gian ƒëƒÉng t·ªõi l∆∞·ª£t t∆∞∆°ng t√°c
```

```python
df_morning = df_full_post[(df_full_post["time"].dt.hour >= 0) & (df_full_post["time"].dt.hour <=11)]
df_afternoon = df_full_post[(df_full_post["time"].dt.hour > 11) & (df_full_post["time"].dt.hour <= 18)]
df_night = df_full_post[(df_full_post["time"].dt.hour > 18)]
values = [df_morning["reaction_count"].mean(), df_afternoon["reaction_count"].mean(), df_night["reaction_count"].mean()]
values_count = [df_morning.shape[0], df_afternoon.shape[0], df_night.shape[0]]
categories = ['morning', 'afternoon', 'night']

plt.figure(figsize=(12, 6))
#bi·ªÅu ƒë·ªì 1
plt.subplot(1,2,1)
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green'])
plt.title("Ph·∫ßn trƒÉm t∆∞∆°ng t√°c trung b√¨nh theo th·ªùi gian ƒëƒÉng\nmorning : 0h-12h\nafternoon : 12h - 19h\nnight : 19h - 0h")
plt.gca().set_aspect('equal')

#bi·ªÅu ƒë·ªì 2
plt.subplot(1, 2, 2)
plt.bar(categories, values_count)
plt.title('S·ªë b√†i post theo th·ªùi gian', pad=20) 
plt.show()
```
